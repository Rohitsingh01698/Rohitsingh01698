```python
defaulter_corr = defaulters.corr()
defaulter_corr_unstck = defaulter_corr.where(np.triu(np.ones(defaulter_corr.shape),k=1).astype(bool)).unstack().reset_index().rename(columns={'level_0':'var1','level_1':'var2',0:'corr'})
defaulter_corr_unstck['corr'] = abs(defaulter_corr_unstck['corr'])
defaulter_corr_unstck.dropna(subset=['corr']).sort_values(by=['corr'],ascending=False).head(10)
```

```python
non_defaulters = non_defaulters.corr()
non_defaulters_unstck = non_defaulters.where(np.triu(np.ones(non_defaulters.shape),k=1).astype(bool)).unstack().reset_index().rename(columns={'level_0':'var1','level_1':'var2',0:'corr'})
non_defaulters_unstck['corr'] = abs(non_defaulters_unstck['corr'])
non_defaulters_unstck.dropna(subset=['corr']).sort_values(by=['corr'],ascending=False).head(10)
```
